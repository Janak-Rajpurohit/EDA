```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("california_cities.csv")
df=df.drop(["Unnamed: 0"],axis=1)  
df.head()
```

```python
plt.scatter(df["latd"],df["longd"],c=np.log10(df["population_total"]),s=df["area_land_km2"],alpha=0.5)
# sns.scatterplot(x=df['latd'],y=df['longd'],hue=np.log10(df["population_total"]))
plt.colorbar(label="log$_{10}$ population")
plt.clim(3,7)
for area in [100,300,500]:
    plt.scatter([],[],c='k',s=area,label=str(area)+'km$^2$',alpha=0.3)
plt.legend(scatterpoints=1,frameon=False,labelspacing=1,title='City Area')
```
