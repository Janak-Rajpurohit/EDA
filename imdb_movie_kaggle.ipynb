```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv("imdb_top_1000.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["Gross"]=df["Gross"].fillna('0')
df["Gross"]=df["Gross"].str.replace(',','')
df["Gross"]=df["Gross"].astype(int)
```

```python
%matplotlib inline
#  top 10 movie with hight votes
df1=df.sort_values(by=["No_of_Votes"],ascending=False)
x=df1["Series_Title"][0:10]
y=df1["No_of_Votes"][0:10]
plt.figure(figsize=(20,7))
plt.title("Top 10 Movies with High no of  votes")
plt.xlabel("Movies")
plt.ylabel("No of Votes")
plt.subplots_adjust(bottom = 0.15)
plt.xticks(rotation=30)
plt.bar(x,y)


```

```python
#  top voted movie with meta score
x=df1["Series_Title"][0:5]
y=df1["Meta_score"][0:5]
plt.xlabel("Movies")
plt.ylabel("MEta Score")
plt.title("top voted movie with meta score")
plt.xticks(rotation=30)
plt.bar(x,y)
```

```python
stars=["Star1","Star2","Star3","Star4"]
```

```python
# Stars with most Occurences in movies

for star in stars:
    x=df[star].value_counts()[0:10].index
    y=df[star].value_counts()[0:10]
    plt.figure(figsize=(15,5))
    plt.xlabel("Stars")
    plt.ylabel("Occurenece")
    plt.xticks(rotation=45)
    plt.title(f"{star}")
    plt.bar(x,y,width=0.5)
    plt.tight_layout()
    
```

```python
#  Tom Hank , Emma Watson , Rupert Grint ,Michael Caine are appeared majority of the times
```

```python
# Apperance of stars in top rated movies
df1=df.sort_values(by=["IMDB_Rating"])
for star in stars:
    plt.figure(figsize=(15,5))
    x=df1[star].value_counts()[0:10].index
    y=df1[star].value_counts()[0:10]
    plt.xlabel("Stars")
    plt.ylabel("Occurence in Top Rated Movies")
    plt.xticks(rotation=45)
    plt.title(f"{star} occurencce in top rated movies")
    plt.bar(x,y,width=0.5)
    plt.tight_layout()
```

```python
# Apperance of stars in top Voted movies
for star in stars:
    df1=df.groupby([star]).sum().reset_index()
    df1=df1.sort_values(by=["No_of_Votes"],ascending=False)
    plt.figure(figsize=(15,5))
    x=df1[star].value_counts()[0:10].index
    y=df1[star].value_counts()[0:10]
    plt.xlabel("Stars")
    plt.ylabel("Occurence in Top voted Movies")
    plt.xticks(rotation=45)
    plt.title(f"{star} occurencce in top voted movies")
    plt.bar(x,y,width=0.5)
    plt.tight_layout()
```

```python
# Apperance of stars in top Gross movies
for star in stars:
    df1=df.groupby([star]).sum().reset_index()
    df1=df1.sort_values(by=["Gross"],ascending=False)
    plt.figure(figsize=(15,5))
    x=df1[star].value_counts()[0:10].index
    y=df1[star].value_counts()[0:10]
    plt.xlabel("Stars")
    plt.ylabel("Occurence in Top Gross Movies")
    plt.xticks(rotation=45)
    plt.title(f"{star} occurencce in top Gross movies")
    plt.bar(x,y,width=0.5)
    plt.tight_layout()
```

```python
# Apperance of stars in top Meta Score movies
for star in stars:
    df1=df.groupby([star]).sum().reset_index()
    df1=df1.sort_values(by=["Meta_score"],ascending=False)
    plt.figure(figsize=(15,5))
    x=df1[star].value_counts()[0:10].index
    y=df1[star].value_counts()[0:10]
    plt.xlabel("Stars")
    plt.ylabel("Occurence in Top Meta Score Movies")
    plt.xticks(rotation=45)
    plt.title(f"{star} occurencce in top Meta Score movies")
    plt.bar(x,y,width=0.5)
    plt.tight_layout()
```

```python
from collections import Counter
genre=[]
for x in df["Genre"]:
    for y in x.split(","):
        genre.append(y.strip().lower())

count= Counter(genre)
count=count.most_common()[0:10]
print(count)
x,y =map(list,zip(*count))
plt.figure(figsize=(20,7))
plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Top Genre")
plt.plot(x,y)
```
